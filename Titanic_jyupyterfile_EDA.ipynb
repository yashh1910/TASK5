```python
# 1. load and inspect the data
# Use Pandas to load your dataset.
import pandas as pd
df = pd.read_csv('train.csv')
```

```python
# Get an overview of the dataset with .info() and .describe().
df.info()   # Data types and missing values
df.describe()  # Summary statistics
```

```python
# Check the distribution of categorical variables with .value_counts()
df['Survived'].value_counts()
```

```python
# 2. Visualize the Data
# Histograms: For numerical features, use histograms to understand the distribution.
import matplotlib.pyplot as plt
df['Pclass'].hist(bins=20)
plt.title('Histogram of Pclass')
plt.show()
```

```python
# 2. Boxplots: Boxplots help identify outliers and the spread of the data.
import seaborn as sns
sns.boxplot(x='Age', data=df)
plt.title('Boxplot of Age')
plt.show()
```

```python
# Pairplot: Use pairplot to explore relationships between numerical variables.
sns.pairplot(df)
plt.show()
```

```python
# Heatmap: For correlations between variables, use a heatmap.
# corr = df.corr()
# sns.heatmap(corr, annot=True, cmap='coolwarm')
# plt.title('Correlation Heatmap')
# plt.show()
# Scatterplots: Explore relationships between two variables with scatterplots.
sns.scatterplot(x='SibSp', y='Parch', data=df)
plt.title('Scatterplot between SibSp and Parch')
plt.show()
```

```python
# 3. Statistical Summary
# Calculate basic statistical measures, such as mean, median, and standard deviation.
df['SibSp'].mean(), df['Parch'].median(), df['Fare'].std()
```
